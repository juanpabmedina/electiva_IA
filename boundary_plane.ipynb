```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = np.random.random_integers(0,10, (50,3))
x_data = data[:,0]
y_data = data[:,1]
z_data = data[:,2]
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x_data, y_data, z_data)
```

```python
def clasificator(test, grades, attitud):
    score = 3*test + 2*grades + attitud - 18 
    if score > 0:
        accepted = 1
    else: 
        accepted = 2
    return accepted
```

```python
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
clasicated_data = []
x = np.arange(0,10)
y = np.arange(0,10)
x, y = np.meshgrid(x, y)
z = - 3*x - 2*y + 18

for x, y, z in data:
   clasicated_data.append(clasificator(x, y, z))
```

```python
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

x = np.arange(0,10)
y = np.arange(0,10)
x, y = np.meshgrid(x, y)
z = - 3*x - 2*y + 18


fig = plt.figure()

ax = fig.add_subplot(projection='3d')
ax.scatter(x_data, y_data, z_data, c=clasicated_data) 
ax.plot_surface(x, y, z, alpha=0.2)

plt.show()
```
