```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Creación de datos aleatorios 
data = np.random.random_integers(0,10, (50,2))
x_data = data[:,0]
y_data = data[:,1]
```

```python
plt.scatter(x_data, y_data)
```

```python
# Frontera
x1 = np.arange(0,10)
x2 = np.arange(0,10)
x2 = -2*x1 + 18
plt.plot(x2)
```

```python
# Función de clasificación
def clasificator(test, grades):
    score = 2*test + grades - 18 
    if score > 0:
        accepted = 1
    else: 
        accepted = 2
    return accepted
```

```python
# Datos clasificados
clasicated_data = []
for x, y in data:
   clasicated_data.append(clasificator(x, y)) 
plt.scatter(x_data, y_data, c=clasicated_data)
x1 = np.arange(0,10)
x2 = -2*x1 + 18
plt.plot(x2)
plt.xlim(-1,11)
plt.ylim(-1,11)
```
